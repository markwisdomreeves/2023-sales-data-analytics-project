Objective: In this project, I performed data analysis on an open dataset that provides an in-depth look at the ordering, billing, and sales processes in a supermarket. Some of the tasks include data cleansing, data transformation (EDA), and data virtualization.

With information ranging from customers' meal choices to the value of their orders and whether they were converted into sales, this dataset opens up endless possibilities to uncover consumer behavior trends and engagement within the business. 

From understanding who is exchanging with the company and when to see what types of meals are most popular with consumers, this rich collection of data will allow us to gain priceless insights into consumer actions and habits that can inform strategic decisions. 

```python
## import libraries 
import pandas as pd 
import numpy as np 
import missingno as msno
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
## Open Dataset link:
## https://zenodo.org/record/4092667#.Y83_G3ZBxhF
```

```python
##import multiple data 
invoive_data=pd.read_csv(r"../supermarket_data_analytics_project/Invoices.csv")
order_data=pd.read_csv(r"../supermarket_data_analytics_project/OrderLeads.csv")
sales_Data=pd.read_csv(r"../supermarket_data_analytics_project/SalesTeam.csv")
```

```python
## convert  all data in  a single data useing  merge function in pandas 
d1=pd.merge(invoive_data,order_data)
d2=pd.merge(d1,sales_Data)
data=d2

# first five rows of data 
data.head()
```

```python
## Checking the size of my dataset 
data.shape[0]
```

```python
## cheaking null value 
data.isnull().sum()
```

```python
## cheaking dtype
data.info()
```

```python
data['Date']=pd.to_datetime(data['Date'], errors='coerce',utc=True)
```

```python
new_data=data[['Participants','Meal Price','Type of Meal','Company Name','Order Value','Converted','Sales Rep','Date']]
```

```python
new_data.info()
```

```python
new_data['day'] = (new_data['Date']).dt.day
new_data['month'] = (new_data['Date']).dt.month
new_data['year'] = (new_data['Date']).dt.year
```

## Exploratory Data Analysis (EDA)

```python
## cheaking the columns 
msno.bar(new_data, figsize = (16,5),color = "magenta")
plt.show()
```

```python
##cheaking relationships 
sns.heatmap(new_data.corr(),annot=True)
```

```python
## now cheak which year has get maximum orders 
new_data['year'].value_counts().plot.bar()
```

```python
## now we cheak which typpe of meal is orderd per year 

print(new_data.groupby('year')['Type of Meal'].value_counts())
print('------------------------------------------------------------')
n_d=new_data[['year','Type of Meal']].value_counts()
data_of_meals_type=pd.DataFrame(n_d,columns=['counts'])
data_of_meals_type.reset_index(inplace=True)
sns.barplot(data=data_of_meals_type, x="year", y="counts", hue="Type of Meal")
```

```python
## now cheak which month has get maximum orders 
new_data['month'].value_counts().plot.bar()
```

```python
## now we cheak which typpe of meal is orderd per month

print(new_data.groupby('month')['Type of Meal'].value_counts())
n_d=new_data[['month','Type of Meal']].value_counts()
data_of_meals_type=pd.DataFrame(n_d,columns=['counts'])
data_of_meals_type.reset_index(inplace=True)
sns.barplot(data=data_of_meals_type, x="month", y="counts", hue="Type of Meal")
```

--> As we seen from above observation that  2018 has highest no of order in it and maximum people has order dinner
--> or as we seen january has maximum nno or order in which most of the people order lunch
